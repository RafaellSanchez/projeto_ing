```python
import sqlite3
```

```python
bd = 'banco_car.db'
banco_caminho = f'/workspaces/projeto_ing/INGESTION/database/{bd}'
conn = sqlite3.connect(banco_caminho)
cursor = conn.cursor()

query = 'select * from tabela_vw'
cursor.execute(query)
resultado = cursor.fetchall()

for linha in resultado:
    print(linha)
```

```python
create_silver = """
create table if not exists tabela_slvr_vw(
    id INTEGER,
    codigo TEXT,
    modelo TEXT,
    dt_igtao DATE
)
"""
cursor.execute(create_silver)
conn.commit()
print('tabela silver criada')

print('schema da tabela silver:')
cursor.execute("PRAGMA table_info(tabela_slvr_vw)")
colunas = cursor.fetchall()
for coluna in colunas:
    print(coluna[1])
```

```python
create_gold = """ 
create table if not exists tabela_gld_vw(
    id INTEGER,
    codigo TEXT,
    modelo TEXT,
    dt_igtao DATE
)
"""
cursor.execute(create_gold)
conn.commit()
print('tabela gold criada')

print('schema da tabela gold:')
cursor.execute("PRAGMA table_info(tabela_gld_vw)")
colunas = cursor.fetchall()
for coluna in colunas:
    print(coluna[1])
```
