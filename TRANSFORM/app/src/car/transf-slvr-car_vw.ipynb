```python
import pandas as pd
import sqlite3
from datetime import datetime
```

```python
bd = 'banco_car.db'
banco_caminho = f'/workspaces/projeto_ing/INGESTION/database/{bd}'
conn = sqlite3.connect(banco_caminho)
cursor = conn.cursor()

query = 'select * from tabela_vw'
cursor.execute(query)
resultado = cursor.fetchall()
```

```python
df = pd.DataFrame(resultado, columns=['codigo', 'modelo', 'id'])
data_igtao = datetime.now().strftime('%Y-%m-%d')
df['dt_igtao'] = data_igtao
print(df.head())
df.to_sql(name='tabela_slvr_vw', con=conn, if_exists = 'append', index = False)
print('tabela ingerida')
```

```python
conn.commit()
```
