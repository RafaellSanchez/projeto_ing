```python
import sqlite3
import pandas as pd
```

```python
bronze = '/workspaces/projeto_ing/database/sqlite/bronze/banco_weather.db'
conn_bronze = sqlite3.connect(bronze)
query_bronze = 'SELECT * FROM tabela_weather'
df_weather_brnz = pd.read_sql_query(query_bronze, conn_bronze)
# display(df_weather_brnz)
df_weather_brnz.dtypes

silver = '/workspaces/projeto_ing/database/sqlite/silver/weather.db'
conn_silver = sqlite3.connect(silver)

for _, row in df_weather_brnz.iterrows():
    query_insert = '''
    INSERT INTO tb_slvr_weather (
        cidade, pais, temperatura, temp_max, temp_min, unidade,prob_chuva, data, latitude, longitude, tempo, dt_igtao
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
    '''
    data = (
        row['cidade'], row['pais'], float(row['temperatura']), float(row['temp_max']), float(row['temp_min']), int(row['unidade']), int(row['prob_chuva']) ,row['data'],
        int(row['latitude']),int(row['longitude']), row['tempo'],
        pd.Timestamp.now().strftime('%Y-%m-%d') 
    )
    conn_silver.execute(query_insert, data)
    
conn_silver.commit()
conn_silver.close()

print("Dados ingeridos com sucesso!")
```
