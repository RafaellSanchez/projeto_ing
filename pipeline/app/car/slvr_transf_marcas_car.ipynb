```python
import sqlite3
import pandas as pd 
```

```python
bronze = '/workspaces/projeto_ing/database/sqlite/bronze/banco_car.db'
conn_bronze = sqlite3.connect(bronze)
query_bronze = 'SELECT * FROM tabela_marcas'
df_marcas_brnz = pd.read_sql_query(query_bronze, conn_bronze)
print(df_marcas_brnz)

conn_bronze.close()

silver = '/workspaces/projeto_ing/database/sqlite/silver/modelos_car.db'
conn_silver = sqlite3.connect(silver)

for _, row in df_marcas_brnz.iterrows():
    query_insert = '''
    INSERT INTO tb_slvr_marcas (
        codigo, modelo, dtIgtao
    ) VALUES (?, ?, ?);
    '''
    data = (
        int(row['id']), row['modelo'],
        pd.Timestamp.now().strftime('%Y-%m-%d') 
    )
    conn_silver.execute(query_insert, data)
conn_silver.commit()
conn_silver.close()

print("Dados ingeridos com sucesso!")
```
