```python
import sqlite3
import pandas as pd
```

```python
bronze = '/workspaces/projeto_ing/database/sqlite/bronze/banco_fakes.db'
conn_bronze = sqlite3.connect(bronze)
query_bronze = 'SELECT * FROM tabela_fakes'
df_fakes_brnz = pd.read_sql_query(query_bronze, conn_bronze)

df_fakes_brnz['cpf'] = df_fakes_brnz['cpf'].str.replace('.', '')
df_fakes_brnz['cpf'] = df_fakes_brnz['cpf'].str.replace('-', '')
df_fakes_brnz['cpf'] = df_fakes_brnz['cpf'].astype(int)
df_fakes_brnz['data_hora'] = pd.Timestamp.now().strftime('%Y-%m-%d:%H:%m:%s')
display(df_fakes_brnz)
df_fakes_brnz.dtypes


conn_bronze.close()

silver = '/workspaces/projeto_ing/database/sqlite/silver/fakes.db'
conn_silver = sqlite3.connect(silver)

for _, row in df_fakes_brnz.iterrows():
    query_insert = '''
    INSERT INTO tb_slvr_fake (
        nome, cpf, email, endereco, bairro, estado, estado_nome, uf, telefone, pais, code,
        cartao_cred, numero_cartao, expr_cartao, cvv_cred, data, data_hora, dt_igtao
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
    '''
    data = (
        row['nome'], int(row['cpf']), row['email'], row['endereco'], row['bairro'], row['estado'], row['estado_nome'], row['sigla'],
        row['telefone'], row['pais'], row['code'], row['cartao_cred'], int(row['cartao_num']),
        row['credt_exp'], int(row['cvv']), row['data'], row['data_hora'],
        pd.Timestamp.now().strftime('%Y-%m-%d') 
    )
    conn_silver.execute(query_insert, data)
conn_silver.commit()
conn_silver.close()

print("Dados ingeridos com sucesso!")
```
