```python
import sqlite3
import pandas as pd
```

```python
bronze = '/workspaces/projeto_ing/database/sqlite/bronze/banco_mega.db'
conn_bronze = sqlite3.connect(bronze)
query_bronze = 'SELECT * FROM tabela_mega'
df_mega_brnz = pd.read_sql_query(query_bronze, conn_bronze)

df_mega_brnz['rateio6_acertos'] = df_mega_brnz['rateio6_acertos'].str.replace('R$', '')
df_mega_brnz['rateio6_acertos'] = df_mega_brnz['rateio6_acertos'].str.replace(',', '')
df_mega_brnz['rateio6_acertos'] = df_mega_brnz['rateio6_acertos'].str.replace('.', '')

df_mega_brnz['rateio5_acertos'] = df_mega_brnz['rateio5_acertos'].str.replace('R$', '')
df_mega_brnz['rateio5_acertos'] = df_mega_brnz['rateio5_acertos'].str.replace(',', '')
df_mega_brnz['rateio5_acertos'] = df_mega_brnz['rateio5_acertos'].str.replace('.', '')

df_mega_brnz['rateio4_acertos'] = df_mega_brnz['rateio4_acertos'].str.replace('R$', '')
df_mega_brnz['rateio4_acertos'] = df_mega_brnz['rateio4_acertos'].str.replace(',', '')
df_mega_brnz['rateio4_acertos'] = df_mega_brnz['rateio4_acertos'].str.replace('.', '')

df_mega_brnz['acumulado6_acertos'] = df_mega_brnz['acumulado6_acertos'].str.replace('R$', '')
df_mega_brnz['acumulado6_acertos'] = df_mega_brnz['acumulado6_acertos'].str.replace(',', '')
df_mega_brnz['acumulado6_acertos'] = df_mega_brnz['acumulado6_acertos'].str.replace('.', '')

df_mega_brnz['estimativa_premio'] = df_mega_brnz['estimativa_premio'].str.replace('R$', '')
df_mega_brnz['estimativa_premio'] = df_mega_brnz['estimativa_premio'].str.replace(',', '')
df_mega_brnz['estimativa_premio'] = df_mega_brnz['estimativa_premio'].str.replace('.', '')

df_mega_brnz['acumulado_mega_virada'] = df_mega_brnz['acumulado_mega_virada'].str.replace('R$', '')
df_mega_brnz['acumulado_mega_virada'] = df_mega_brnz['acumulado_mega_virada'].str.replace(',', '')
df_mega_brnz['acumulado_mega_virada'] = df_mega_brnz['acumulado_mega_virada'].str.replace('.', '')

df_mega_brnz['arrecadacao_toral'] = df_mega_brnz['arrecadacao_toral'].str.replace('R$', '')
df_mega_brnz['arrecadacao_toral'] = df_mega_brnz['arrecadacao_toral'].str.replace(',', '')
df_mega_brnz['arrecadacao_toral'] = df_mega_brnz['arrecadacao_toral'].str.replace('.', '')

df_mega_brnz['rateio6_acertos'] = df_mega_brnz['rateio6_acertos'].astype(int)
df_mega_brnz['rateio5_acertos'] = df_mega_brnz['rateio5_acertos'].astype(int)
df_mega_brnz['rateio4_acertos'] = df_mega_brnz['rateio4_acertos'].astype(int)

df_mega_brnz['acumulado6_acertos'] = df_mega_brnz['acumulado6_acertos'].astype(int)
df_mega_brnz['estimativa_premio'] = df_mega_brnz['estimativa_premio'].astype(int)
df_mega_brnz['acumulado_mega_virada'] = df_mega_brnz['acumulado_mega_virada'].astype(int)
df_mega_brnz['arrecadacao_toral'] = df_mega_brnz['arrecadacao_toral'].astype(int)

display(df_mega_brnz)

conn_bronze.close()

silver = '/workspaces/projeto_ing/database/sqlite/silver/mega.db'
conn_silver = sqlite3.connect(silver)

for _, row in df_mega_brnz.iterrows():
    query_insert = '''
    INSERT INTO tb_slvr_mega (
        concurso, data_sorteio, bola1, bola2, bola3, bola4, bola5, bola6, ganhadores6_acertos, cidade_uf, rateio6,
        ganhadores5_acertos, rateio5, ganhadores4_acertos, rateio4, acumulado6_acertos, arrecadacao_total, estimativa_premio, acumulado_mega_virada,
        observacao, dt_igtao
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
    '''
    data = (
        row['Concurso'], row['data_sorteio'], int(row['bola1']), int(row['bola2']), int(row['bola3']), int(row['bola4']), int(row['bola5']), int(row['bola6']),
        int(row['ganhadores6_acertos']), row['cidade_uf'], int(row['rateio6_acertos']), int(row['ganhadores5_acertos']), int(row['rateio5_acertos']),
        int(row['ganhadores4_acertos']), int(row['rateio4_acertos']), int(row['acumulado6_acertos']), int(row['arrecadacao_toral']), int(row['estimativa_premio']),
        int(row['acumulado_mega_virada']), row['observacao'],
        pd.Timestamp.now().strftime('%Y-%m-%d') 
    )
    conn_silver.execute(query_insert, data)
conn_silver.commit()
conn_silver.close()

print("Dados ingeridos com sucesso!")
```
