```python
import sqlite3
import pandas as pd 
```

```python
bronze = '/workspaces/projeto_ing/database/sqlite/bronze/banco_calendario.db'
conn_bronze = sqlite3.connect(bronze)
query_bronze = 'SELECT * FROM tabela_calendario'
df_calendar_brnz = pd.read_sql_query(query_bronze, conn_bronze)
print(df_calendar_brnz)

conn_bronze.close()

silver = '/workspaces/projeto_ing/database/sqlite/silver/calendario.db'
conn_silver = sqlite3.connect(silver)

for _, row in df_calendar_brnz.iterrows():
    query_insert = '''
    INSERT INTO tb_slvr_celendar_feri (
        nomeFeriado, descricao, dataFeri, ano, mes, dia, dtIgtao
    ) VALUES (?, ?, ?, ?, ?, ?, ?);
    '''
    data = (
        row['nome'], row['descricao'], row['data_ISO'],
        row['ano'], row['mes'], row['dia'],
        pd.Timestamp.now().strftime('%Y-%m-%d') 
    )
    conn_silver.execute(query_insert, data)
conn_silver.commit()
conn_silver.close()

print("Dados ingeridos com sucesso!")
```
